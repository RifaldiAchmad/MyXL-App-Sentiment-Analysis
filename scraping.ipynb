```python
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
import csv
```

```python
# Mengambil semua ulasan dari aplikasi di Google Play Store
scrapreview = reviews_all(
    'com.taxsee.taxsee',          # ID aplikasi
    lang='id',             # Bahasa ulasan (default: 'en')
    country='id',          # Negara (default: 'us')
    sort=Sort.MOST_RELEVANT, # Urutan ulasan (default: Sort.MOST_RELEVANT)
    count=1000
)
```

```python
# Filter ulasan yang memiliki 'content' valid (tidak None atau string kosong)
valid_reviews = [review for review in scrapreview if review.get('content')]
```

```python
df = pd.DataFrame(valid_reviews)
df.head()
```

```python
from datetime import datetime

# Ubah kolom 'at' menjadi objek datetime jika belum dalam format datetime
if not pd.api.types.is_datetime64_any_dtype(df['at']):
    df['at'] = pd.to_datetime(df['at'])

# Filter DataFrame berdasarkan rentang tanggal yang ditentukan
start_date = datetime(2024, 3, 1)
end_date = datetime(2025, 3, 1)

training_df = df[(df['at'] >= start_date) & (df['at'] <= end_date)]
training_df = training_df.sort_values(by='at', ascending=False)
training_df.head()
```

```python
training_df.to_excel('maxim_training.xlsx', index=False)
```
